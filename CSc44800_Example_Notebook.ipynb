## The goal of the next block is to display the results of given equations in a loop and displaying the results in a table. 

```python
import math
import pandas

result = pandas.DataFrame(columns = ['k', 'x', '((e^x)-1/x)','((e^x)-1/log(e^x)'])
pandas.set_option('float_format', '{:f}'.format)

for k in range(1,16):
    x = 10**(-1*k)
    f1 = ((math.exp(x))-1)/x
    f2 = (math.exp(x)-1)/(math.log(math.exp(x)))
    result.loc[k] = [k, x, f1, f2]

print(result)
```

What this code does specifically is that it first initializes result as a table using DataFrame with column names we find appropriate.

pandas.set_option proceeds to make the results from each equation a float because otherwise it would've been automatically rounded up.

In the loop, we wish to see the results of the numbers in a range of 1 to 15 when plugged into the two given equations. During each iteration of the loop, we proceed to enter the current values of the variables into the table.

Finally, we print out the table in all its glory.

## Quantiative vs Quantitative Graph using Vegalite

```python
from vega import VegaLite

VegaLite(
{
    "$schema": "https://vega.github.io/schema/vega-lite/v5.json",
    "description": "Graph 1 for Vega lite practice. Quantitative vs Quantitative",
    "data": {"url": "https://raw.githubusercontent.com/Primagon/Visualization-Vega-lite-practice/main/imports-85.csv"},
    "mark": "point",
    "encoding": {
        "x": {"field": "city-mpg", "type": "quantitative"},
        "y": {"field": "price", "type": "quantitative"}
    }
}
)
```

This segment is directly extracted from the vegalite editor using the Scatterplot prompt from: https://vega.github.io/vega-lite/examples/point_2d.html

The different is that we changed the description to fit the current dataset.
The data where we extracted from is from my personal github.
And lastly, the fields are different to fit the columns we wish to see a relationship between.
Because of the values in "city-mpg" and "price" we realize the type is quantitative because both columns have only numerical values.

## Quantitatve vs Categorical using Vegalite

```python
VegaLite(
{
    "$schema": "https://vega.github.io/schema/vega-lite/v5.json",
    "description": "Graph 2 for Vega lite practice. Quantitative vs Categorical",
    "data": {"url": "https://raw.githubusercontent.com/Primagon/Visualization-Vega-lite-practice/main/imports-85.csv"},
    "mark": "point",
    "encoding": {
        "x": {"field": "make", "type": "nominal"},
        "y": {"field": "price", "type": "quantitative"}
    }
}
)
```

The different is that we changed the description to fit the current dataset.
The data where we extracted from is from my personal github.
And lastly, the fields are different to fit the columns we wish to see a relationship between.
Because of the values in "make" column, we must change the type from quantitative to nominal because it is a string and has no ranking to it so therefore it is not ordinal.
